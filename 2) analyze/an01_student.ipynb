```python
import pandas as pd

def clean(file):
    # 1열은 필요없으므로 건너뛴 상태로 csv파일 부름
    F = pd.read_csv(file, skiprows = 1)       

    # 형변환
    F['특성별(1)'] = F['특성별(1)'].astype('category')
    F['특성별(2)'] = F['특성별(2)'].astype('category')

    # 멀티인덱스 설정
    F.set_index(['특성별(1)','특성별(2)'], inplace = True)

    # 정보 확인
    F.info()

    # 들어온 파라미터의 원본을 최종적으로 저장
    file = F

    return file

# --------------------------------------------------------------------------
# 함수 이름 : select
# 매개 변수 : F
# 함수 기능 : 데이터 프레임에서 해당 인덱스 열을 뽑아줌 (DataFrame 형식)
# --------------------------------------------------------------------------
def select(F, indexC):
        result = F.xs(indexC, level = 0)
        return result

# --------------------------------------------------------------------------
# 함수 이름 : ana
# 매개 변수 : name
# 함수 기능 : 각 프레임을 보여주고 describe한 결과를 보여줌
# --------------------------------------------------------------------------
def ana(name):
    print(name)
    print()
    print(name.T.describe())
```

```python
file2 = '../data_student/학생피해_설문2_가장_두려운_사이버폭력_유형_123순위.csv'
file2 = clean(file2)
```

```python
file2
```

```python
# 필요한 데이터프레임만 선택하여 분리함
#성별, 학령별, 학교특성별, 거주지역별, 인터넷 이용 시간별, 가정형편별, 학업성적별, 사이버폭력 예방교육별

gender = select(file2, '성별')
school = select(file2, '학령별')
schooltype = select(file2, '학교특성별')
area = select(file2, '거주지역별')
hour = select(file2, '인터넷 이용 시간별')
family = select(file2, '가정형편별')
score = select(file2, '학업성적별')
edu = select(file2, '사이버폭력 예방교육별')
```

### 각 프레임에서 가장 두려워하는 사이버폭력 유형이 무엇인지 분석

```python
ana(gender)
#성별에 상관없이 사이버명예훼손
```

```python
ana(school)
#사이버명예훼손이 모두 두려움
```

```python
ana(schooltype)
#사이버명예훼손이 모두 두려움
```

```python
ana(area)
#사이버명예훼손이 모두 두려움
```

```python
ana(hour)
#사이버명예훼손이 모두 두려움
```

```python
ana(family)
#상위와 중위는 사이버 명예훼손
#하위는 신상정보 유출
```

```python
ana(score)
#사이버명에훼손
```

```python
ana(edu)
#경험유무와 상관없이 사이버명예훼손
```
