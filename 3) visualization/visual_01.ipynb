# 성인_피해_1) 가장 두려운 사이버 폭력 유형

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl                 # 패키지 관련
import matplotlib.font_manager as fm 
import numpy as np


def clean(file):
    # 1열은 필요없으므로 건너뛴 상태로 csv파일 부름
    F = pd.read_csv(file, skiprows = 1)       

    # 형변환
    F['특성별(1)'] = F['특성별(1)'].astype('category')
    F['특성별(2)'] = F['특성별(2)'].astype('category')

    # 멀티인덱스 설정
    F.set_index(['특성별(1)','특성별(2)'], inplace = True)

    # 정보 확인
    F.info()

    # 들어온 파라미터의 원본을 최종적으로 저장
    file = F

    return file


# --------------------------------------------------------------------------
# 함수 이름 : select
# 매개 변수 : F (데이터 프레임 변수 이름), indexC (행 인덱스 이름)
# 함수 기능 : 데이터 프레임에서 해당 인덱스 열을 뽑아줌 (DataFrame 형식)
# --------------------------------------------------------------------------
def select(F, indexC):
        result = F.xs(indexC, level = 0)
        return result

```

```python
# 전체 크기 설정
plt.figure(figsize=(14,10))

xData = ['사이버명예훼손', '사이버언어폭력', '사이버성폭력', '사이버스토킹', '신상정보유출', '사이버따돌림', '사이버갈취', '사이버강요']
valueData = [56, 50.3, 49.6, 47.4, 42.5, 23, 18.2, 12.9]

plt.bar(range(len(valueData)), valueData, tick_label = xData, width = 0.5 )
# tick_label = x축 라벨
# width = 각 막대의 폭

plt.show()
```

```python
# 파일을 불러옴
file2 = '../data_adult/성인피해_설문2_가장_두려운_사이버폭력_유형_123순위.csv'
file2 = clean(file2)
```

```python
# 필요한 데이터프레임만 선택
gender = select(file2, '성별')
age = select(file2, '연령별')
degree = select(file2, '학력별')
job = select(file2, '직업별')
area = select(file2, '거주지역별')
hour = select(file2, '인터넷 이용시간별')
edu = select(file2, '사이버폭력예방교육별')
```

```python
def graph1_adult(data_name, x):
    # 전체 크기 설정
    plt.figure(figsize=(15,10))
    # 각 데이터 프레임 별로 시각화를 함
    graph = data_name.plot(kind = 'bar', grid = True, xlabel = x , ylabel ='백분위', rot = 10, fontsize = 'small')
    graph = plt.xlabel(x, loc = 'right', fontsize = 'large')
    graph = plt.ylabel('백분위', loc = 'top', fontsize = 'large')
    graph = plt.title('[성인 - 집단별 가장 두려운 사이버 폭력 유형]')
    graph = plt.legend(['사이버명예훼손', '사이버언어폭력', '사이버성폭력', '사이버스토킹',
                        '신상정보유출', '사이버따돌림', '사이버갈취', '사이버강요'], fontsize = 'xx-small')
    
```

```python
graph1_adult(gender, '성별')
graph1_adult(age, '연령별')
graph1_adult(degree, '학력별')
graph1_adult(job, '직업별')
graph1_adult(area, '거주지역별')
graph1_adult(hour, '인터넷 이용시간별')
graph1_adult(edu, '사이버폭력 예방교육별')
```

# 학생

```python
def graph1_stu(data_name, x):
    # 전체 크기 설정
    plt.figure(figsize=(15,10))
    # 각 데이터 프레임 별로 시각화를 함
    graph = data_name.plot(kind = 'bar', grid = True, xlabel = x , ylabel ='백분위', rot = 10, fontsize = 'small')
    graph = plt.xlabel(x, loc = 'right', fontsize = 'large')
    graph = plt.ylabel('백분위', loc = 'top', fontsize = 'large')
    graph = plt.title('[학생 - 집단별 가장 두려운 사이버 폭력 유형]')
    graph = plt.legend(['사이버 명예훼손', '신상정보 유출', '사이버 언어폭력', '사이버 따돌림', '사이버 성폭력',
                        '사이버 스토킹', '사이버 강요', '사이버 갈취'], fontsize = 'xx-small')
    
```

```python
file2 = '../data_student/학생피해_설문2_가장_두려운_사이버폭력_유형_123순위.csv'
file2 = clean(file2)

# 필요한 데이터프레임만 선택
gender = select(file2, '성별')
school = select(file2, '학령별')
area = select(file2, '거주지역별')
hour = select(file2, '인터넷 이용 시간별')
family = select(file2, '가정형편별')
score = select(file2, '학업성적별')
edu = select(file2, '사이버폭력 예방교육별')
```

```python
# 그래프 그리기
graph1_stu(gender, '성별')
graph1_stu(school, '학령별')
graph1_stu(area, '거주지역별')
graph1_stu(hour, '인터넷 이용 시간별')
graph1_stu(family,'가정형편별')
graph1_stu(score, '학업성적별')
graph1_stu(edu, '사이버 폭력 에방교육별')
```
