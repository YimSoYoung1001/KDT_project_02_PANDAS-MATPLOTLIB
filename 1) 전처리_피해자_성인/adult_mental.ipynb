```python
import pandas as pd
```

```python
file1 = '../data_adult/성인피해_사이버폭력_피해_경험_후_심리상태__1사람들을_만나거나_사귀기가_힘들었음.csv'
# 파일들은 쉼표로 구분됨
# 각 value들은 %에 해당함. 그래서 한 행을 모두 합하면 100이 나옴
```

```python
file1 = pd.read_csv(file1, skiprows = 1)
file1

# csv 파일을 읽어올 때 1번행까지 skip해서 가져오겠다.
```

```python
file1.info()
```

```python
file1['특성별(1)'] = file1['특성별(1)'].astype('category')
file1['특성별(2)'] = file1['특성별(2)'].astype('category')
file1.info()
```

```python
# 열 0번과 1번을 멀티인덱스로 원본에 설정
file1.set_index(['특성별(1)','특성별(2)'], inplace = True)
file1

# 참고
# 멀티인덱스로 설정할 때 열이름만 넣어야한다.
# 만약 [file1['특성별(1)'],file1['특성별(2)']] 이르케 한다면 그거는 복붙한 결과 나옴
```

```python
file1.iloc[1]
```

```python
# ==========================================================
# 함수 이름 : clean
# 함수 기능 : 행과 열 인덱스를 맞춤
# 변     수 : 파일을 지정해야하여 변수로 설정해야함
# ==========================================================

def clean(file):
    # 1열은 필요없으므로 건너뛴 상태로 csv파일 부름
    F = pd.read_csv(file, skiprows = 1)       

    # 형변환
    F['특성별(1)'] = F['특성별(1)'].astype('category')
    F['특성별(2)'] = F['특성별(2)'].astype('category')

    # 멀티인덱스 설정
    F.set_index(['특성별(1)','특성별(2)'], inplace = True)

    # 정보 확인
    F.info()                

    return F
```

```python
file2 = '../data_adult/성인피해_사이버폭력_피해_경험_후_심리상태__2평소에_하던_일_업무가_하기_싫어졌음.csv'
file2 = clean(file2)
```
